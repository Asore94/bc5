```python
Months = {'January':1, 'February':2, 'March':3, 'April':4, 'May':5, 'June':6, 'July':7, 'August':8, 'September':9, 'October':10, 'November':11, 'December':12}
df['Date'] = df.Year.astype(str) + '-' + df.Month.map(Months).astype(str).str.zfill(2) + '-' + df.Day.astype(str).str.zfill(2)
df.head()
```

```python
df.to_parquet('Data/Daily/', engine='pyarrow', compression='snappy', index=False, partition_cols=['Date'])
```

```python
import glob

dates_to_read = ['2019-01-01', '2019-01-02']

list_of_files = []
newdf_data = []

for day in dates_to_read:
    list_of_files.append(glob.glob(f'./Data/Daily/Date={day}/*.parquet'))

print(list_of_files)

for item in list_of_files:
    content = pd.read_parquet(item, columns=['Year', 'Month', 'Day', 'Station', 'Covid19'])
    newdf_data.append(content)

new_df = pd.concat(newdf_data)

new_df.tail()
```
